1) Utilize the datasets in your Pandas exercise. Read and merge the two datasets.

2) Univariate analysis is the simplest form of data analysis where the data being analyzed contains
only one variable. Follow the instructions and explain your findings.

    a) Create a pie chart to show the percentage of Marital/Family status.

        Also, include the following:

            i) Figure title, Slice labels and Label percentage

    
    b) Create a histogram to show frequency distributions of applicants income.
        
        Also, include the following:
            i) Figure title, Labels, Grid and change le6 tick label style = plain

```python
# Create a single Visualization

family_status_counts = merged_df['NAME_FAMILY_STATUS'].value_counts()

# Plot the pie chart
plt.figure(figsize=(10, 6))
plt.pie(family_status_counts, labels=family_status_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Family Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

According to the data, 70.3% of the data are married.  

```python
# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(merged_df['AMT_INCOME_TOTAL'], bins=10, edgecolor='black')

# Add title and labels
plt.title('Frequency Distribution of Applicants\' Income')
plt.xlabel('Income')
plt.ylabel('Frequency')

# Change le6 tick label style to plain
plt.ticklabel_format(style='plain', axis='x')
```

While majority of the data are married, majority of the data have an income of below 200000 pesos.

3) Bivariate analysis is used to determine the relationship between two variables. Follow the
instructions and explain your findings.

    a) Create a scatter plot to observe relationships between family members count and
    children count. Also, include the following:

    i) Figure title, Labels, Marker = * , and cmap = Spectral

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['CNT_FAM_MEMBERS'], merged_df['CNT_CHILDREN'], marker='*', cmap='Spectral')

# Add title and labels
plt.title('Relationship Between Family Members Count and Children Count')
plt.xlabel('Family Members Count')
plt.ylabel('Children Count')

# Show the plot
plt.show()
```

The scatterplot shows that there is a linear trend upwards that indicates a positive correlation between children count and family members

```python
# Use preprocessed
df2 = pd.read_csv("ccapp_record.csv")
```

```python
#Install seaborn
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df2.corr()

# Create the correlation heatmap with annotations
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')

# Add title
plt.title('Correlation Map of Different Variables')

```

The warmer the color is the more closely related the variable is from each other. 

5) Identifying outliers and missing data through visualization. Follow the instructions and explain
your findings.

    a) Create a boxplot to show outliers in income, family members count, and children. Use 1 figure with 3 axes. Also, include the following:
        
        i) Figure and Axes Titles and outliers points color: income = red, family size = green, and children = blue
    
    b) Create a horizontal bar chart to show missing values in your dataset. Also, include the following:
        
        i) Figure title and bar color = orange

```python
# Create one figure with 3 subplots (axes) for the boxplot
fig, axes = plt.subplots(1, 3, figsize=(15, 6))
fig.suptitle('Boxplots Showing Outliers in Various Features')

# Boxplot for income
axes[0].boxplot(merged_df['AMT_INCOME_TOTAL'], vert=True, patch_artist=True,
                flierprops={'markerfacecolor': 'r', 'marker': 'o', 'markersize': 8})
axes[0].set_title('Income')
axes[0].set_xticklabels('')

# Boxplot for family members count
axes[1].boxplot(merged_df['CNT_FAM_MEMBERS'], vert=True, patch_artist=True,
                flierprops={'markerfacecolor': 'g', 'marker': 'o', 'markersize': 8})
axes[1].set_title('Family Members Count')
axes[1].set_xticklabels('')

# Boxplot for children count
axes[2].boxplot(df2['CNT_CHILDREN'], vert=True, patch_artist=True,
                flierprops={'markerfacecolor': 'b', 'marker': 'o', 'markersize': 8})
axes[2].set_title('Children Count')
axes[2].set_xticklabels('')
```

```python
# Calculate the number of missing values for each column
missing_values_count = merged_df.isnull().sum()

# Create a horizontal bar chart to show missing values
plt.figure(figsize=(10, 6))
plt.barh(missing_values_count.index, missing_values_count.values, color='orange')

# Add title and labels
plt.title('Missing Values in Dataset')
plt.xlabel('Number of Missing Values')
plt.ylabel('Features')
```

Before preprocessing, there are missing data present from the merged df, specifically in the occupation type column.
