7) It is apparent that the columns/features DAYS_BIRTH and DAYS_EMPLOYED contain negative
values. It is your task to write a function that converts/computes DAYS_BIRTH into age in years
and DAYS_EMPLOYED into years in employment. Also, remove the columns
MONTHS_BALANCE and STATUS.

```python
merged_df.head()
```

```python
dropped_merged = merged_df.drop(columns=['MONTHS_BALANCE', 'STATUS'])
dropped_merged.head()
```

```python
# Convert columns DAYS_BIRTH and DAYS_EMPLOYED to positive
dropped_merged['DAYS_BIRTH'] = dropped_merged['DAYS_BIRTH'].abs()
dropped_merged['DAYS_EMPLOYED'] = dropped_merged['DAYS_EMPLOYED'].abs()
dropped_merged.head()
```

```python
# Convert to years by dividing to 365
dropped_merged['DAYS_BIRTH'] = dropped_merged['DAYS_BIRTH'] / 365
dropped_merged['DAYS_EMPLOYED'] = dropped_merged['DAYS_EMPLOYED']  / 365
dropped_merged.head()
```

```python
%pip install numpy
```

```python
import numpy as np
```

```python
# Round down to avoid decimal place
dropped_merged['DAYS_BIRTH'] = np.floor(dropped_merged['DAYS_BIRTH'])
dropped_merged['DAYS_EMPLOYED'] = np.floor(dropped_merged['DAYS_EMPLOYED'])
dropped_merged.head()
```
